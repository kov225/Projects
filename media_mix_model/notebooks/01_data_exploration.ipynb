# 📈 MMM Project: Exploratory Data Analysis

This notebook explores the simulated marketing telemetry for our DTC brand. We examine spend trends, channel correlations (multicollinearity), and the relationship between marketing inputs and total conversions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
palette = sns.color_palette("muted")

# Load data
df = pd.read_csv('../data/simulated_data.csv')
df.head()
```

### 1. Spend Distributions across Channels

We visualize how spend is distributed over the 156-week period. Note the Q4 seasonality peaks.

```python
media_cols = ['linear_tv_spend', 'ctv_spend', 'paid_search_spend', 'paid_social_spend', 'display_spend']
df[media_cols].plot(figsize=(14, 6), title='Weekly Marketing Spend by Channel')
plt.ylabel('Spend ($)')
plt.show()
```

### 2. Multi-collinearity Analysis

High correlation between Linear TV and CTV spend (r=0.6) presents a challenge for frequentist models, which we address through Bayesian priors.

```python
corr = df[media_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spend Correlation Matrix')
plt.show()
```

### 3. Conversion Dynamics

Total website conversions exhibit clear seasonality and growth trends.

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='week', y='conversions', color='darkblue')
plt.title('Weekly Website Conversions')
plt.show()
```
