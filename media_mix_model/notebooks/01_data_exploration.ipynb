# MMM Data Exploration

Initial visualization of simulated DTC brand telemetry. We analyze spend correlations, seasonality, and organic trends.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/simulated_data.csv')
df.head()
```

```
plt.figure(figsize=(12, 6))
plt.plot(df['week'], df['conversions'], label='Weekly Conversions')
plt.title('Weekly Website Conversions (3 Years)')
plt.legend()
plt.show()
```
