# 📡 TV Attribution: Traffic and Airings Exploratory Analysis

In this notebook, we visualize minute-level website telemetry, identify organic seasonality, and overlay the TV ad airing log to understand the baseline traffic environment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')

# Load data
sessions = pd.read_csv('../data/sessions.csv', parse_dates=['timestamp'])
airings = pd.read_csv('../data/airings.csv', parse_dates=['timestamp'])

sessions.head()
```

### 1. Minute-Level Traffic Baseline

We visualize the first 24 hours of traffic to see the daily organic peaks and baseline noise floor.

```python
day_1 = sessions[:1440]
plt.figure(figsize=(15, 6))
plt.plot(day_1['timestamp'], day_1['sessions'], color='darkcyan', label='Website Sessions')
plt.title('Minute-Level Sessions Baseline (Day 1)')
plt.ylabel('Sessions')
plt.legend()
plt.show()
```

### 2. Airing Log Distribution

We overlay the 300 ad airings onto the 90-day time series to visualize the coverage across networks.

```python
plt.figure(figsize=(15, 4))
plt.scatter(airings['timestamp'], airings['network'], alpha=0.5, s=20, color='red')
plt.title('TV Ad Airing Log (90 Days)')
plt.xlabel('Date')
plt.ylabel('Network')
plt.show()
```

### 3. Control Market Correlation

The control market (zero TV advertising) provides the necessary covariate for our causal impact model. Correlation: r=0.85.

```python
control = pd.read_csv('../data/control_market.csv', parse_dates=['timestamp'])
merged = pd.merge(sessions, control, on='timestamp', suffixes=('_treated', '_control'))
merged[:1440].plot(x='timestamp', title='Treated vs Control Market (Organic Correlation)')
plt.show()
```
