# TV Attribution: Traffic and Airings Exploration

Visualizing minute-level web telemetry and TV airing logs.

```
import pandas as pd
import matplotlib.pyplot as plt

sessions = pd.read_csv('../data/sessions.csv', parse_dates=['timestamp'])
airings = pd.read_csv('../data/airings.csv', parse_dates=['timestamp'])

plt.figure(figsize=(15, 5))
plt.plot(sessions['timestamp'][:1440], sessions['sessions'][:1440])
plt.title('Minute-Level Baseline Traffic (Day 1)')
plt.show()
```
